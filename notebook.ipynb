<h2> is there any null values present in any column?!! , visualize the result </h2>

```python
#.isnull()==>

# to show where the null value is present 
data.isnull()

# to count the null values in each column
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt


mylabels = list(data.isnull())

wedgeprops = {"linewidth": 4, 'width':1, "edgecolor":"k"} # Width = 1
textprops = {"fontsize":16}# Font size of text in pie chart
plt.pie(data.isnull().sum(),labels=mylabels, 
        autopct = "%0.2f%%", # Show data in persentage for with 2 decimal point
        shadow = True, # Showing shadow of pie chart
        radius = 3.55, # Radius to increase or decrease the size of pie chart 
       startangle = 270, # Start angle of first section
       counterclock = True, 
        wedgeprops = wedgeprops,
        textprops =textprops,
        labeldistance = 1.1
       
        
        
        )
plt.legend()
plt.figure(figsize=(270,280))
plt.show() 
```
